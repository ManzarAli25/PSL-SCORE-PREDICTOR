```python
import numpy as np
import pandas as pd
from yaml import safe_load
import os
from tqdm import tqdm
import pickle
```

```python
first_innings = pickle.load(open('first_innings.pkl','rb'))
matches = pickle.load(open('dataset_level1.pkl','rb'))

```

```python
first_innings['innings']='1st'
```

```python
first_innings
```

```python
count = 1
delivery_df_2 = pd.DataFrame()
for index, row in matches.iterrows():
    if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count+=1
        continue
    count+=1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    # Check the length of the row['innings'] list
    if len(row['innings']) > 1:
        # Only proceed if there is a second innings
        for ball in row['innings'][1]['2nd innings']['deliveries']:
            for key in ball.keys():
                match_id.append(count)
                batting_team.append(row['innings'][1]['2nd innings']['team'])
                teams.append(row['info.teams'])
                ball_of_match.append(key)
                batsman.append(ball[key]['batsman'])
                bowler.append(ball[key]['bowler'])
                runs.append(ball[key]['runs']['total'])
                city.append(row['info.city'])
                venue.append(row['info.venue'])
                try:
        
                    player_of_dismissed.append(ball[key]['wicket']['player_out'])
                except:
                    player_of_dismissed.append('0')
        loop_df = pd.DataFrame({
                'match_id':match_id,
                'teams':teams,
                'batting_team':batting_team,
                'ball':ball_of_match,
                'batsman':batsman,
                'bowler':bowler,
                'runs':runs,
                'player_dismissed':player_of_dismissed,
                'city':city,
                'venue':venue
            })
        delivery_df_2 = pd.concat([delivery_df_2,loop_df],ignore_index=True)
```

```python
delivery_df_2
```

```python
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
delivery_df_2['bowling_team'] = delivery_df_2.apply(bowl,axis=1)

#added an aditional column
```

```python
delivery_df_2.drop('teams', axis=1, inplace=True)

delivery_df_2
```

```python
first_df = delivery_df_2[['match_id','batting_team','bowling_team','ball','runs','player_dismissed','city','venue']]
first_df
```

```python

if first_df['city'].isnull().any():
    first_df.loc[first_df['city'].isnull(), 'city'] = first_df.loc[first_df['city'].isnull(), 'venue'].str.split().str[0]
first_df['city'].unique()
```

```python
x = np.where(first_df['city'].isnull(), first_df['venue'].str.split().apply(lambda x:x[0]), first_df['city'])
```

```python
first_df['city']=x
```

```python
first_df.drop(columns=['venue'],inplace=True)
```

```python

first_df['runs'] = pd.to_numeric(first_df['runs'], errors='coerce')

# Group by 'match_id' and calculate cumulative sum for 'runs'
first_df['current_score'] = first_df.groupby('match_id')['runs'].cumsum()
```

```python
first_df['over'] = first_df['ball'].apply(lambda x:str(x).split(".")[0])
first_df['ball_no'] = first_df['ball'].apply(lambda x:str(x).split(".")[1])
```

```python
first_df['balls_bowled'] = (first_df['over'].astype('int')*6) + first_df['ball_no'].astype('int')
```

```python
first_df['balls_left'] = 120 - first_df['balls_bowled']

first_df['balls_left'] = first_df['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python

# Convert 'player_dismissed' column to numeric dtype
first_df['player_dismissed'] = pd.to_numeric(first_df['player_dismissed'], errors='coerce')

first_df['player_dismissed'] = first_df['player_dismissed'].apply(lambda x: 1 if x != 0 else x)

# Group by 'match_id' and calculate cumulative sum for 'player_dismissed'
first_df['wickets_left'] = 10 - (first_df.groupby('match_id')['player_dismissed'].cumsum())


```

```python
first_df['crr'] = (first_df['current_score']*6)/first_df['balls_bowled']
```

```python
first_df['runs'] = pd.to_numeric(first_df['runs'], errors='coerce')

groups = first_df.groupby('match_id')

match_ids = first_df['match_id'].unique()
last_five = []

for match_id in match_ids:
    # Get the group and calculate rolling sum for 'runs' column
    group_df = groups.get_group(match_id)
    rolling_sum = group_df['runs'].rolling(window=30, min_periods=1).sum()
    
    # Extend the list with rolling sum values
    last_five.extend(rolling_sum.values.tolist())

# Add 'last_five' column to the DataFrame
first_df['last_five'] = last_five
```

```python
first_df.loc[first_df['balls_bowled'] <= 29, 'last_five'] = 0
```

```python
final_df = first_df.groupby('match_id').sum()['runs'].reset_index().merge(first_df,on='match_id')
```

```python
final_df.rename(columns={'runs_x': 'total_score'}, inplace=True)
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','total_score']]

```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df['innings']='2nd'
```

```python
second_innings = final_df

pickle.dump(final_df,open('second_innings.pkl','wb'))
```

```python
both_innings = pd.concat([first_innings, second_innings], ignore_index=True)
```

```python
X = both_innings.drop(columns=['total_score'])
y = both_innings['total_score']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city','innings'])
]
,remainder='passthrough')
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.1,max_depth=10,random_state=42))
])
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print("r2 score is, ",r2_score(y_test,y_pred))
print("mean absolute error is ",mean_absolute_error(y_test,y_pred))
```
